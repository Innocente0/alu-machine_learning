```python
#!/usr/bin/env python3
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec

y = np.arange(0, 11) ** 3

#Defining X
x = np.arange(0,11)
y = x ** 3

# Plot
plt.figure(figsize=(6,4))
plt.plot(x, y, marker='o', linestyle='-', color='red')
plt.title('Cube Function: y = x\u00b3')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#!/usr/bin/env python3

y0 = np.arange(0, 11) ** 3

mean = [69, 0]
cov = [[15, 8], [8, 15]]
np.random.seed(5)
x1, y1 = np.random.multivariate_normal(mean, cov, 2000).T
y1 += 180

x2 = np.arange(0, 28651, 5730)
r2 = np.log(0.5)
t2 = 5730
y2 = np.exp((r2 / t2) * x2)

x3 = np.arange(0, 21000, 1000)
r3 = np.log(0.5)
t31 = 5730
t32 = 1600
y31 = np.exp((r3 / t31) * x3)
y32 = np.exp((r3 / t32) * x3)

np.random.seed(5)
student_grades = np.random.normal(68, 15, 50)

# your code here
# ————— Plotting All 5 in One Figure —————
fig = plt.figure(figsize=(10, 10))
fig.suptitle('All in One', fontsize='x-small')

gs = gridspec.GridSpec(3, 2, figure=fig, hspace=0.4, wspace=0.3)

# 1) y = x^3
ax0 = fig.add_subplot(gs[0, 0])
ax0.plot(np.arange(0, 11), y0)
ax0.set_title('y = x³', fontsize='x-small')
ax0.set_xlabel('x', fontsize='x-small')
ax0.set_ylabel('y', fontsize='x-small')

# 2) Scatter: Height vs Weight
ax1 = fig.add_subplot(gs[0, 1])
ax1.scatter(x1, y1, color='magenta')
ax1.set_title("Men's Height vs Weight", fontsize='x-small')
ax1.set_xlabel('Height (in)', fontsize='x-small')
ax1.set_ylabel('Weight (lbs)', fontsize='x-small')

# 3) Exponential Decay of C-14
ax2 = fig.add_subplot(gs[1, 0])
ax2.plot(x2, y2)
ax2.set_title('Exponential Decay of C-14', fontsize='x-small')
ax2.set_xlabel('Time (years)', fontsize='x-small')
ax2.set_ylabel('Fraction Remaining', fontsize='x-small')
ax2.set_xbound(0, 28650)
ax2.set_yscale('log')

# 4) Two Radioactive Decays
ax3 = fig.add_subplot(gs[1, 1])
ax3.plot(x3, y31, 'r--', label='C-14')
ax3.plot(x3, y32, 'g-',  label='Ra-226')
ax3.set_title('Exponential Decay of Radioactive Elements', fontsize='x-small')
ax3.set_xlabel('Time (years)', fontsize='x-small')
ax3.set_ylabel('Fraction Remaining', fontsize='x-small')
ax3.set_xlim(0, 20000)
ax3.set_ylim(0, 1)
ax3.legend(loc='upper right', fontsize='x-small')

# 5) Histogram of Grades
ax4 = fig.add_subplot(gs[2, :])  # spans both columns
bins = np.arange(0, 101, 10)
ax4.hist(student_grades, bins=bins, edgecolor='black', color='skyblue')
ax4.set_title('Project A', fontsize='x-small')
ax4.set_xlabel('Grades', fontsize='x-small')
ax4.set_ylabel('Number of Students', fontsize='x-small')
ax4.set_xticks(bins)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
#!/usr/bin/env python3

np.random.seed(5)
fruit = np.random.randint(0, 20, (4,3))

# your code here
people = np.arange(fruit.shape[1])
names = ['Farrah', 'Fred', 'Felicia']
fruit_names = ['Apples', 'Bananas', 'Oranges', 'Peaches']
colors = ['red', 'yellow', '#ff8000', '#ffe5b4']

# Plotting stacked bars
plt.figure(figsize=(6, 4))
bottom = np.zeros(fruit.shape[1])  # start bottoms at 0

for idx in range(fruit.shape[0]):
    plt.bar(
        people,
        fruit[idx],
        bottom=bottom,
        width=0.5,
        color=colors[idx],
        label=fruit_names[idx]
    )
    bottom += fruit[idx]  # update bottom for next stack

# Labels, title, legend, and ticks
plt.xticks(people, names)
plt.ylabel('Quantity of Fruit')
plt.ylim(0, 80)
plt.yticks(np.arange(0, 81, 10))
plt.title('Number of Fruit per Person')
plt.legend(title='Fruit')

plt.tight_layout()
plt.show()
```
