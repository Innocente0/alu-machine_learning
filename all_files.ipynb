```python
#!/usr/bin/env python3
import numpy as np
import matplotlib.pyplot as plt

y = np.arange(0, 11) ** 3

#Defining X
x = np.arange(0,11)
y = x ** 3

# Plot
plt.figure(figsize=(6,4))
plt.plot(x, y, marker='o', linestyle='-', color='red')
plt.title('Cube Function: y = x\u00b3')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#!/usr/bin/env python3
import numpy as np
import matplotlib.pyplot as plt

mean = [69, 0]
cov = [[15, 8], [8, 15]]
np.random.seed(5)
x, y = np.random.multivariate_normal(mean, cov, 2000).T
y += 180

# your code here
plt.figure(figsize=(8, 6)) # You can adjust the figure size as needed
plt.scatter(x, y, s=5, alpha=0.5, color='magenta') # Use scatter plot for individual points
plt.title('Scatter Plot of Generated Data')
plt.xlabel('Height(in)')
plt.ylabel('Weight(lbs)')
plt.grid(True) # Added  grid for better readability
plt.show()
```
